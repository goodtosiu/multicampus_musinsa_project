```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from datetime import datetime

# =========================
# 0. ë¡œë”© ë¡œê·¸
# =========================
def loading(msg):
    print(f"[LOADING] {msg}")

def now():
    return datetime.now().strftime("%H:%M:%S")

# =========================
# 1. Selenium ì„¤ì •
# =========================
options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 25)

loading("ë¸Œë¼ìš°ì € ì‹¤í–‰")

# =========================
# 2. ë¬´ì‹ ì‚¬ ì ‘ì†
# =========================
driver.get("https://www.musinsa.com/main/musinsa/recommend?gf=M&tabId=brand")
time.sleep(5)
loading("ë¬´ì‹ ì‚¬ ì ‘ì† ì™„ë£Œ")

# =========================
# 3. Joyride íˆ´íŒ ì œê±°
# =========================
def close_joyride_tooltip():
    try:
        btns = driver.find_elements(
            By.CSS_SELECTOR,
            'button[aria-label="ë‹«ê¸°"], button[data-action="primary"]'
        )
        for btn in btns:
            try:
                driver.execute_script("arguments[0].click();", btn)
                time.sleep(0.4)
            except:
                continue
    except:
        pass

close_joyride_tooltip()
time.sleep(1)
close_joyride_tooltip()

# =========================
# 4. ë¸Œëœë“œ ë©”ë‰´ íƒ­ í´ë¦­
# =========================
brand_tabs = wait.until(
    EC.presence_of_all_elements_located(
        (By.CSS_SELECTOR, 'span[data-mds="TabTextItem"]')
    )
)

for tab in brand_tabs:
    if tab.text.strip().startswith("ë¸Œëœë“œ"):
        driver.execute_script("arguments[0].click();", tab)
        loading("ë¸Œëœë“œ ë©”ë‰´ íƒ­ í´ë¦­")
        break

time.sleep(2)

# =========================
# 5. ë¸Œëœë“œ ë©”ë‰´ ìŠ¤í¬ë¡¤ ì»¨í…Œì´ë„ˆ
# =========================
scroll_container = wait.until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, 'div[data-scroll-id="menu-sub-area"]')
    )
)
loading("ë¸Œëœë“œ ë©”ë‰´ ë¡œë”© ì™„ë£Œ")

# =========================
# 6. íŒ¨ì…˜ì†Œí’ˆ ë²„íŠ¼ í´ë¦­
# =========================
fashion_btn = wait.until(
    EC.element_to_be_clickable((
        By.XPATH,
        '//div[@data-button-name="íŒ¨ì…˜ì†Œí’ˆ"]'
    ))
)
driver.execute_script("arguments[0].click();", fashion_btn)
loading("íŒ¨ì…˜ì†Œí’ˆ ì¹´í…Œê³ ë¦¬ ì„ íƒ")
time.sleep(2)

# =========================
# 7. ë¸Œëœë“œ ì „ì²´ ìˆ˜ì§‘ (LIMIT ì—†ìŒ)
# =========================
brand_data = {}

loading("ë¸Œëœë“œ ì „ì²´ ìˆ˜ì§‘ ì‹œì‘ (ë¬´í•œ ìŠ¤í¬ë¡¤)")

while True:
    brand_items = driver.find_elements(
        By.CSS_SELECTOR,
        'a[href*="/brand/"] div.UIBrandItem__NameWrap-sc-1un5tti-4'
    )

    for name_wrap in brand_items:
        try:
            brand_link = name_wrap.find_element(By.XPATH, './ancestor::a')
            brand_url = brand_link.get_attribute("href")

            brand_kr = name_wrap.find_element(
                By.CSS_SELECTOR,
                'span.UIBrandItem__Name-sc-1un5tti-6'
            ).text.strip()

            try:
                brand_en = name_wrap.find_element(
                    By.CSS_SELECTOR,
                    'span.UIBrandItem__EnglishName-sc-1un5tti-7'
                ).text.strip()
            except:
                brand_en = None

            if brand_url and brand_url not in brand_data:
                brand_data[brand_url] = {
                    "brand_kr": brand_kr,
                    "brand_en": brand_en,
                    "brand_url": brand_url
                }

                # ì¸ë±ìŠ¤ + ì‹œê°„ ë¡œê·¸
                print(f"[{now()}] index={len(brand_data)} | {brand_kr} | {brand_url}")

        except:
            continue

    # ğŸ”¥ virtual listëŠ” ë°˜ë“œì‹œ ì¡°ê¸ˆì”© ë‚´ë ¤ì•¼ í•¨
    driver.execute_script(
        "arguments[0].scrollTop += 120;",
        scroll_container
    )
    time.sleep(0.6)

# =========================
# 8. (Ctrl+C ì´í›„) ê²°ê³¼ ì €ì¥
# =========================
# ìˆ˜ë™ ì¤‘ë‹¨ í›„ ì•„ë˜ ì‹¤í–‰
# df = pd.DataFrame(brand_data.values())
# df.to_csv("musinsa_fashion_goods_brands_all.csv", index=False, encoding="utf-8-sig")
```
